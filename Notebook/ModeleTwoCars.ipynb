### Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the parameters 
a2:="$\alpha_2$" <br>
h:="time step for the discretization" <br>
T:="the Time of the simulation

```python
# Define the parameters
a2 = 2.0  #capacity of acceleration deceleration
h = 1.0  # Time step
T = 10.0  # Total simulation time
n_steps = int(T / h)  # Number of time steps
```

### Initialisztion for the position of the first car, the second car (behind the first car) and the time step table

```python
# Initialize arrays to store the results
x1Pos = np.zeros(n_steps)
x2Pos = np.zeros(n_steps)
time = np.zeros(n_steps)
```

### Define the initial condition
position of the first car, the second car behind the first car and t0=0

```python
# Initial conditions
time[0] = 0.0  # Initial time
x1Pos[0] = 2.0  # Initial value for x1 as specified
x2Pos[0] = 1.0  # Initial value for x2 as specified
```

### Definition of $x_1'$ and $x'_2$
$$
\begin{align*}
x'_1 &= V_{\text{max}_1} = 130 \text{ km/h} \\
x'_2 &= \alpha_2(x_1-x_2)
\end{align*}
$$


```python
def x2_prime(t):
    return a2 * (x1Pos[t] - x2Pos[t])

def x1_prime():
    return 130 * (1000 / 3)
```

### Applying Euler's Method for the Resolution of the EDO
The idea is to loop on each time step. Because the speed of the first car is constant, the position increases linearly with the following formula:
$$
\begin{align*}
&x_1(t + \Delta t) = x_1 + V_{\text{max}_1} \cdot h \\
&\text{And by Euler's method:} \\
&x_2(t + \Delta t) = x_2(t) + h \cdot \alpha_2(x_1(t)-x_2(t)) \end{align*}$$


```python
# Euler's explicit method
for t in range(1, n_steps):
    
    x1Pos[t] = x1Pos[t-1] + x1_prime() * h
    x2Pos[t] = x2Pos[t-1] + x2_prime(t-1) * h
    time[t] = time[t-1] + h
```

### Finally plot the result


```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, x1Pos, label='x1(t)')
plt.plot(time, x2Pos, label='x2(t)')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.title('accordion effect')
plt.grid(True)
plt.show()
```
